```python
# Import necessary library for data manipulation
import pandas as pd

# Load the CSV into a Pandas DataFrame
df = pd.read_csv('sample_data.csv')

# Display the whole DataFrame
df

# Display the shape of the DataFrame (number of rows, number of columns)
df.shape

# Display the first 5 rows of the DataFrame
df.head()

# Display the first 10 rows of the DataFrame
df.head(10)

# Display the first 15 rows of the DataFrame
df.head(15)

# Display the last 5 rows of the DataFrame
df.tail()

# Display the last 10 rows of the DataFrame
df.tail(10)

# Display the last 15 rows of the DataFrame
df.tail(15)

# Print the column names of the DataFrame
df.columns

# Display concise summary of the DataFrame, including data types and missing values
df.info()

# Display descriptive statistics of the DataFrame
df.describe

# Calculate the total number of elements in the DataFrame
df.size

# Display data types of each column
df.dtypes

# Display the number of dimensions (axes) of the DataFrame
df.ndim

# Check for missing values in the entire DataFrame
df.isnull()

# Check for missing values in each column and sum them up
df.isnull().sum()

# Check if there are any missing values in the entire DataFrame
df.isnull().values.any()

# Check if there are missing values in the 'Country' column
df['Country'].isnull()

# Fill missing values in the 'Age' column with the median value
df['Age'].fillna(value=df.Age.median(), inplace=True)
df

# Fill missing values in 'Salary' with the median, 'Country' with 'France', and 'Purchased' with 'Yes'
df['Salary'].fillna(value=df.Salary.median(), inplace=True)
df['Country'].fillna(value='France', inplace=True)
df['Purchased'].fillna(value='Yes', inplace=True)
df

# Display the median value of the 'Age' column
print("Fill the missing values.")
median = df["Age"].median()
print(median)

# Fill missing values in the 'Age' column with the median
df["Age"].fillna(median, inplace=True)
print(df.describe())
print("\n")
print(df)

# Drop rows with any missing values
df.dropna(inplace=True)
df.shape

# Drop duplicate rows from the DataFrame
df = df.drop_duplicates()
print(df)
df.shape

# Print the 'Country' column for the first 6 rows
print(df[0:6]['Country'])

# Print the 'Country' and 'Age' columns for all rows
print(df.loc[:, ['Country', 'Age']])

# Print the 'Country' and 'Age' columns for rows 0 to 6
print(df.loc[0:6, ['Country', 'Age']])

# Print the 'Country' and 'Age' columns for specific rows (1, 3, 5)
print(df.loc[[1, 3, 5], ['Country', 'Age']])
```
